```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import re
import numpy as np
import pandas as pd
```

```python
df_for_youtube = df_for_youtube[['Comment']]
```

```python
df_for_youtube.head()
```

```python
# Instantiate Model

tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
def sentiment_score(review):
  tokens = tokenizer.encode(review, return_tensors='pt')
  result = model(tokens)
  return int(torch.argmax(result.logits))+1
```

## For chapa_kata_trailer_comments from youtube

```python
df_chapa_kata_trailer_comments = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ML_project/datasets/chapa_kata_trailer_comments.csv')
df_chapa_kata_trailer_comments.head()
```
